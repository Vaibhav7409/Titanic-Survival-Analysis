```python
#importing libraries 
import pandas as pd
import matplotlib as plt
import seaborn as sns

#importing data
df = pd.read_excel("C:/Users/vsing/Downloads/Titanic.xlsx", sheet_name="Titanic.csv")
df.shape
df.info()
df.describe()
```

```python
#dropping unnecessary columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df = df.drop(['PassengerId', 'Ticket', 'Cabin', 'Name'], axis=1)
```

```python
df.info()
```

```python
#finding missing values
df.isnull().sum()
```

```python
# Fill Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill Embarked with most common value (mode)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.head()
```

```python
# 1. Create FamilySize column
df['FamilySize'] = df['SibSp'] + df['Parch']

# 2. AgeGroup based on Age
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 17, 60, 120], labels=['Child', 'Adult', 'Senior'])

# 3. Survival Status (for readability)
df['SurvivalStatus'] = df['Survived'].map({1: 'Survived', 0: 'Died'})
```
